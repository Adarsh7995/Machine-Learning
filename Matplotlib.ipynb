```python
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline
```

# Matplotlib figure/graph anatomy

```python
#Create data and its function
x=np.linspace(0.,10.,100)
y=x*x
```

```python
#Show and save your line plot
plt.figure(figsize=(5,5))
plt.plot(x,y,label='$x^2$')
plt.title('Square Function')
plt.legend(loc='best')
plt.xlabel('x')
plt.ylabel('$x^2$')
```

```python
#Create data and its function
x=np.linspace(0. ,10. ,100)
y=np.sin(x)#in radian
```

```python
plt.figure(figsize=(9,6))
plt.plot(x,y,'r--',label='sin(x)')
plt.title('Sinusoidal Function')
plt.legend(loc= 'best')
plt.xlabel('Phase Angle (rad)')
plt.ylabel('sin(x)')
plt.savefig('sin_function.png',dpi=300)
```

```python
items=np.array(['coke','pepsi','fanta','maaza','mirinda'])
qty=np.array([100,85,20,30,45])
plt.bar(items,qty)
plt.title("Sales")
plt.xlabel('Beverages')
plt.ylabel('Qty sold')
plt.show()
```

```python
plt.pie(qty,labels=items,autopct='%0.1f')
plt.title('% of sales')
plt.show()
```

```python
x=np.linspace(0,10,100)
fig=plt.figure()
plt.plot(x,np.sin(x),'-')
plt.plot(x,np.cos(x),'--');
```

# Subplots

```python
plt.figure()
# create a plot figure
# create the first of two panels and set current axis
plt.subplot(2,1,1)
# (rows, columns, panel number)
plt.plot(x,np.sin(x))
plt.title('sin(x)')
plt.ylabel('sin(x)')
# create the second panel and set current axis
plt.subplot(2,1,2)
plt.plot(x,np.cos(x));
plt.title('cos(x)')
plt.ylabel('cos(x)')
plt.xlabel('x')
```

```python
plt.figure()# create a plot figure
fig,ax=plt.subplots(2,2)
plt.tight_layout()
# Call plot() method on the appropriate object
ax[0,0].plot(x,np.sin(x))
ax[0,0].set_title('Sin(x)')
ax[0,0].set_ylabel('sin(x)')
ax[0,1].plot(x,np.cos(x))
ax[1,0].plot(x,np.tan(x))
ax[1,1].plot(x,np.exp(x))
plt.setp(ax[-1, :],xlabel='x')
plt.setp(ax[:,0],ylabel='f(x)')
```
