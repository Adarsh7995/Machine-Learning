```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([[1,6],[1,4],[1,8],[1,6],[1,7]])
y=np.array([65,70,60,52,69])
y=y.reshape(-1,1)
theta=np.array([0.5,1])
alpha=0.005
```

```python
plt.scatter(x[:,1],y)
```

```python
for epo in range(20):
    for i in range(len(x)):
        y_cap=np.dot(theta,x[i])
        e=y_cap-y[i]
        grad=e*x[i]
        theta=theta-alpha*grad
print(theta)
px=np.linspace(2,10,25)
py=theta[0]+theta[1]*px
plt.plot(px,py)
plt.scatter(x[:,1],y)
```
